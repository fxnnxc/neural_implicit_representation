```python
import os 
from PIL import Image
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
path = "Set14"
images = [] 
names = sorted(os.listdir(path))
for image_name in names:
    img = Image.open(os.path.join(path, image_name))
    img = np.array(img)
    images.append(img)
    print(f"{image_name:15s} | shape:{img.shape} | max:{img.max()} | min:{img.min()}")
```

```python
half = len(images)//2
fig, ax = plt.subplots(2, half, figsize=(15,8))
for i, img in enumerate(images):
    ax[i//half, i%half].imshow(img)
```
