```python
import os 
from PIL import Image
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
path = "Set5"
images = [] 
names = sorted(os.listdir(path))
for image_name in names:
    img = Image.open(os.path.join(path, image_name))
    img = np.array(img)
    images.append(img)
    print(f"{image_name:15s} | shape:{img.shape} | max:{img.max()} | min:{img.min()}")
```

```python
fig, ax = plt.subplots(1, len(images), figsize=(15,4))
for i, img in enumerate(images):
    ax[i].imshow(img)
```
